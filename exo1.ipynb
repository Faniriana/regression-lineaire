```python
import numpy as np
import matplotlib.pyplot as plt

# Génération des données
np.random.seed(42)
x = np.random.uniform(0, 10, 100)
y = 2 * x + 5 + np.random.normal(0, 1, 100)  # Ajout d'un bruit aléatoire

# Initialisation des paramètres
m = np.random.randn()  # pente initiale
c = np.random.randn()  # intercept initial
alpha = 0.01  # taux d'apprentissage
epochs = 1000  # nombre d'itérations

# Implémentation de la descente de gradient
n = len(x)
for epoch in range(epochs):
    # Prédictions
    y_pred = m * x + c
    
    # Calcul des gradients
    dm = -(2/n) * np.sum(x * (y - y_pred))
    dc = -(2/n) * np.sum(y - y_pred)
    
    # Mise à jour des paramètres
    m -= alpha * dm
    c -= alpha * dc
    
    # Calcul de la perte
    loss = np.mean((y - y_pred) ** 2)
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

# Résultats finaux
print(f"Paramètres finaux : m = {m:.4f}, c = {c:.4f}")

# Visualisation
plt.scatter(x, y, color='blue', label='Données originales')
plt.plot(x, m * x + c, color='red', label='Ajustement')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
